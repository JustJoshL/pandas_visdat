### 7. Menghapus missing values pada DataFrame

```python
import pandas as pd
import numpy as np  
print(pd.__version__)
print(np.__version__)  
```

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, np.nan, 3, np.nan, 5],
    'D': [1, 2, 3, 4, 5]
}).reset_index()

df.head()
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)  # Membuat salinan cadangan DataFrame
```

```python
#menghapus kolom missing values pada DataFrame
df = df.dropna(axis='columns')  # Menghapus kolom dengan missing values
df.head()
```

```python
#Menghapus setiap baris yang mengandung missing values
df = df_backup.copy(deep=True)  # Mengembalikan DataFrame dari cadangan
df = df.dropna(axis='rows')
df.head()
```

```python
#Persentase missing values untuk tiap kolom
df = df_backup.copy(deep=True)
df.isna().mean()
```

```python
#menghapus setiap kolom yang memiliki missing values berdasarkan threshold
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

### 8. Memeriksa kesamaan antar kolom/series pada Data Frame

```python
data = {'A': [15, 15, 18, np.nan, 12],
        'B': [15, 15, 18, np.nan, 12],}

df = pd.DataFrame(data)
df
```

```python
#Mengenal pandas series
df['A']
```

```python
type(df['A'])
```

```python
type(df)
```

```python
#Memeriksa kesamaan dengan operator ==
df['A'] == df['B'] #value "nan" tidak bisa membandingkan dengan baik
```

```python
#Alternatif: .equals(), membandingkan dua buah kolom lebih baik menggunakan ini.
df['A'].equals(df['B'])
```

```python
#.equals() atau == bisa digunakan antar dua DataFrame
df1 = df.copy(deep=True)
df.equals(df1)
```
