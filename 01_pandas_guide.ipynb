## Pemilihan Baris

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns=cols)
df
```

Fungsi Operator Logika |

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

Fungsi isin()

```python
df[(df['A'].isin([1,3]))]
```

Operator Negasi "~"

```python
df[~df['A'].isin([1, 3])]
```

## Konversi String ke Numerik

```python
data = {'col1':['1', '2', '3', 'teks'],
        'col2':['1', '2', '3', '4']}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

Konversi tipe data dengan fungsi astype()

```python
df_x = df.astype({'col2':'float'})
df_x
```

```python
df_x.dtypes
```

```python
df.apply(pd.to_numeric, errors='coerce')
```

## Pemilihan Kolom Berdasarkan Tipe Data

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float')

df.index = pd.date_range(start='2025-01-01', periods=n_rows, freq='H')
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
df.dtypes
```

Tipe data Numerik

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='float')
```

```python
df.select_dtypes(include='int')
```

String/Object (tipe datanya)

```python
df.select_dtypes(include='object')
```

datetime tipe data

```python
df.select_dtypes(include='datetime')
```

kombinasi tipe data

```python
df.select_dtypes(include=['number', 'object'])
```
