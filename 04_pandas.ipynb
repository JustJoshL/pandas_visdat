### 13. Konversi nilai numerik ke dalam kategori pada DataFrame

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), columns=cols)
df
```

Pengelompokan nilai numerik berupa usia menggunakan cut()

```python
df['kelompok_usia'] = pd.cut(df['usia'], 
                            bins=[0, 18, 65, 99],
                            labels = ['anak', 'dewasa', 'manula'])
df
```

### 14. Merge 2 Pandas DF

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

```python
#Menggunakan Method pd.merge() inner = data yang ada di kedua df
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
#Outer merge = semua data dari kedua df (yang ada duplikasinya diadopsi di salah satu aja)
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

### 15. Memecah nilai string suatu kolom ke dalam beberapa kolom baru di DF

```python
data = {'nama': ['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'],
        'tempat_kelahiran': ['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = pd.DataFrame(data)
df
```

```python
#mau misahin nama depan dan blakang (Expand = mengizinkan hasil split jadi beberapa kolom)
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

```python
#sama untuk tempat lahir
df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

### 16. Menggunakan unstack() pada multiple indexes di DF

menata ulang data framenya yang ada multiple index

```python
df = pd.read_csv('https://raw.githubusercontent.com/boedybios/kaggle_explorations/refs/heads/master/belajar_matplotlib_dasar/data/titanicfull.csv')
df.head()
```

```python
#yang mau diamati itu survivednya, dikelompokkan dalam female dan male, yang diakses hanya aggregate value dari survivednya aja, dalam bentuk rata".
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame() #frame itu buat mempercantik layoutnya.
```

```python
#nah kalo ada multiple index, kita bisa atur ulang pake unstack() (biar si tulisan survivednya turun sejajar ke barisan sex dan pclass)
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()
```
