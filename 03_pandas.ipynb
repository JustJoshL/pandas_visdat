### 10. Mengganti nama kolom pada Data Frame berdasarkan pola

```python
df = pd.read_csv('https://raw.githubusercontent.com/boedybios/kaggle_explorations/refs/heads/master/belajar_matplotlib_dasar/data/titanicfull.csv')
df.columns = ['Pclass', 'Survival status', 'full Name', 'Sex ', ' Age', 'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
df_backup =df.copy(deep=True)

df.head()
```

```python
#menggunakan lowercase untuk nama kolom & mengganti spasi dengan "_"
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

```python
#Memangkas kelebihan spasi pada nama kolom
df = df_backup.copy(deep=True)  #kembalikan df ke kondisi awal
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_') #str.strip akan membuang spasi yang berlebihan (di awal age atau setelah sex)
df.head()
```

### 11. Seleksi kolom dan baris pada Data Frame menggunakan loc

```python
n_rows= 10
n_cols= 5
cols= tuple('ABCDE')
df= pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df
```

```python
#Seleksi kolom dan baris pake loc
df.loc[[0,3,4], ['B','E']] #kiri: baris, kanan: kolom
```

```python
#Seleksi dengan kondisi
df.loc[df['B']>10, ['B','D','E']]
```

```python
df.loc[0:4, 'B':'D'] #teknik slicing, kolom B sampai D, baris 0 sampai 4
#sifatnya inklusif untuk slicingnya, semuanya di include.
```

### 12. Membentuk kolom bertipe datetime dari beberapa kolom lain.

```python
data = {
    'day': [1, 2, 10, 25, 12],
    'month': [1, 2, 4, 5, 6],
    'year': [2000, 2001, 2010, 2015, 2020]}
df = pd.DataFrame(data)
df
```

```python
df['penanggalan'] = pd.to_datetime(df[['day', 'month', 'year']]) #nama kolom gaboleh sembarangan. ga bisa pake bahasa lain, harus day, month, year
df
```
