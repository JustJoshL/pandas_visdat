### 53. Membuat kategori baru menurut threshold (ambang batas)

```python
d = {'hobi':['jogging', 'mancing', 'renang',
             'mancing', 'mancing', 'baca', 
             'baca', 'mancing', 'fotografi',
             'mancing', 'camping']}

df = pd.DataFrame(d)
df
```

```python
df['hobi'].value_counts()
```

```python
persentase = df['hobi'].value_counts(normalize=True)
persentase
```

```python
threshold = 0.1
hobi_lain = persentase[persentase < threshold].index
hobi_lain
```

```python
df['hobi'] = df['hobi'].replace(hobi_lain, 'lainnya')
df
```

```python
df['hobi'].value_counts(normalize=True) 
#Model/Studi kasus ini sering ditemui dalam data science. Proporsi kecil, gabungkan menjadi besar.
```

### 55. Menghitung jumlah baris menurut kriteria tertentu

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
#Menghitung jumlah baris menurut kriteria tertentu
(df['A'] < 5).sum()
```

```python
#Proporsi terhadap keseluruhan data frame
(df['A'] < 5).mean()
```
